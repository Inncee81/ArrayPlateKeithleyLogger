#### Brown 2004 - The Physics of Ion Sources. Page 62 Child-Langmuir law limited emission current.

```python
import numpy as np

r = 0.010/2
d = 0.006
U = 8000
e0 = 8.854e-12
q = 1.602e-19
m = 39.948*1.661e-27
S = r/d
I = 4/9*np.pi*e0*(2*q/m)**0.5*S**2*U**(3/2)
print(f'Max. current = {I*1000:.2f} mA')
```

```python
import matplotlib.pyplot as plt
r = np.arange(0.000,0.008,0.0005)
I = 4/9*np.pi*8.854e-12*(2*1.602e-19/(39.948*1.661e-27))**0.5*(r/d)**2 * U**(3/2)
plt.figure(figsize = (8,3), dpi=276)
plt.plot(r*2000,I*1000,'.')
plt.xlabel('Aperture Diameter (mm)')
plt.ylabel('Max. Current (mA)')
plt.grid()
plt.show()
```
